```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[100,101,102,103,104]

plt.plot(x,y);
plt.title("First Plotting")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

students=['Arin','Mahima','Neelesh','Parth']
marks=[99,25,90,48]

plt.bar(students,marks,color='skyblue')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student-Marks Comparision")
plt.show()
```

```python
import matplotlib.pyplot as plt

height=[5,5.5,5.8,6.1]
weight=[55,60,69,74]

plt.scatter(height,weight,color='skyblue')
plt.title("Height vs Weight Comparison")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
from matplotlib import markers
import matplotlib.pyplot as plt

days=[1,2,3,4,5]
temp=[25,27.5,30,32.5,35]
humidity=[20,25,30,35,40]

plt.plot(days,temp,label='Temperature(C)',color='red',marker='o')
plt.plot(days,humidity,label='Humidity(%)',color='blue',marker='s')
plt.title("Weather Report")
plt.xlabel("Days")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=[0,10,100]
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.plot(x,y1,color='red')
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,y2,color='blue')
plt.title("Cosine")

plt.tight_layout()
plt.show()
```
